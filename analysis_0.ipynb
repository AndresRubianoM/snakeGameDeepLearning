```python
games = sampler(SNAKE_GAMES[0], REWARDS[0], 0, 0, 300, True)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

for game in games:
    x = range(len(game))
    ax.plot(x, game, alpha=0.2)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))

for game in games[0:3]:
    x = range(len(game)-1)
    game_diff = []
    for i in x:
        game_diff.append(game[i+1]-game[i])
        
    ax.plot(x, game_diff)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
last = []
for game in games:
    last.append(game[-1])

ax.hist(last, bins =30)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
last = []
for game in games:
    last.append(len(game))

ax.hist(last, bins =30)


plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
last = []
duration = []
for game in games:
    last.append(game[-1])
    duration.append(len(game))

ax.hist2d(duration, last,bins =40)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))

for game in games:
    x = range(len(game))
    game_diff = []
    for i in range(len(game)-1):
        game_diff.append(game[i+1]-game[i])
    game_diff = np.array([0] + game_diff)
    game_ = np.array(game)
    result = game_*game_diff
    ax.scatter(x, result, alpha=0.15)

plt.show()
```
